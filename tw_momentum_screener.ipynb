## 步驟 2：載入程式

```python
import tw_momentum_screener as tw
from tw_momentum_screener import CFG

# 可在此調整參數（選用）
CFG.confidence_threshold = 70   # 信心閥值，越高越嚴格
CFG.max_picks = 20
CFG.top_n = 3
CFG.use_full_universe = False   # True = 掃描全上市櫃（較慢）；False = 內建精選股票池
CFG.fetch_sentiment = True      # 是否抓新聞情緒（只對前 30 名候選）
CFG.use_industry = True         # 族群（產業）相對強度，抓族群帶頭股
CFG.use_finmind = False         # 回測時用 FinMind 抓歷史法人（讓回測納入籌碼面）
# CFG.finmind_token = "你的token"  # 選填，提高 FinMind 額度
```

## 步驟 3：每日盤後選股
